**INPUT DATA**

```
import numpy as np
import matplotlib.pyplot as plt

path = "data.csv"
data = np.genfromtxt(path, delimiter=',')

x_data = data[:, 0]
y_data = data[:, 1]

plt.figure(figsize=(8, 8))
plt.scatter(x_data, y_data, alpha=0.3)
plt.show()
```

**LINEAR REGRESSION RESULTS**

```
from sklearn import linear_model

x_data = x_data.reshape(-1, 1)
y_data = y_data.reshape(-1, 1)

train_data = linear_model.LinearRegression()
train_data.fit(x_data, y_data)
plt.figure(figsize=(8, 8))
plt.scatter(x_data, y_data, color='black')
plt.plot(x_data, train_data.predict(x_data), color='red')
plt.show()
```

**ENERGY SURFACE**

```
fig = plt.figure(figsize=(7, 7))
energy_surface = fig.add_subplot(projection='3d')
theta0_axis = np.arange(-30, 30, 0.1)
theta1_axis = np.arange(-30, 30, 0.1)
THETA0, THETA1 = np.meshgrid(theta0_axis, theta1_axis)
totalError = 0
for i in range(len(x_data)):
  totalError += (y_data[i] - (THETA0 * x_data[i] + THETA1)) ** 2 / (float(len(x_data)) * 2)

energy_surface.plot_surface(THETA0, THETA1, totalError, cmap = 'rainbow', alpha=0.8)
energy_surface.set_xlabel('Theta 0')
energy_surface.set_ylabel('Theta 1')
energy_surface.set_zlabel('Error')
energy_surface.view_init(35,40)
```
